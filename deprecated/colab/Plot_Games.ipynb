
#3D Sufaces, Contours, and Heatmap Fun
This is a starter snippet for plotting 3D Surfaces, Contour lines, and heatmaps. Let's first import necessary libraries


```
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

Define an x,y grid. linspace is very hand for that. Let's generate a sequence of a 100 numbers between -2 and 2 for both x and y axis.

```
xlist = np.linspace(-3, 3, 100)
ylist = np.linspace(-3, 3, 100)
```

Let's define the full grid. Numpy.meshgrid is great since it will generate all the pair of points using the x, and y intervals

```
X, Y = np.meshgrid(xlist, ylist)
```

Finally, let's calculate some function. I choose a joint probability of 2 Random Variables with standard normal distributions.

```
W = 1/(2*np.pi)*np.exp(-X**2/2)*np.exp(-Y**2/2)
```

First comes the surface plot

```
fig = plt.figure()
ax = Axes3D(fig)

ax.plot_surface(X, Y, W)

ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

Let's compare that with a contour plot.

```
plt.contour(X, Y, W)
plt.show()
```

Finally, a heatmap

```
sns.heatmap(data=W, annot=False, cmap="magma", xticklabels=False, yticklabels=False)
plt.show()
```
